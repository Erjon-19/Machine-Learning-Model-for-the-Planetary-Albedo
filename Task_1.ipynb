```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data from url
Albedo = pd.read_csv('https://raw.githubusercontent.com/ML4SCI/ML4SCI_GSoC/main/Messenger/Moon/Albedo_Map.csv', header=None)
```

```python
# check what kind of data are 
Albedo.shape
```

```python
Albedo.describe()
```

```python
# reproducing the maps
# 720/360=2 so the aspect ratio of figure is 2  
plt.figure(figsize=(12,6))
sns.heatmap(Albedo, cmap="gray", xticklabels=False, yticklabels=False)
```

```python
# loadind the other maps into pandas DataFrames
#iron map potassium, thorium, titanium map
LPFe = pd.read_csv('https://raw.githubusercontent.com/ML4SCI/ML4SCI_GSoC/main/Messenger/Moon/LPFe_Map.csv', header=None)
LPK = pd.read_csv('https://raw.githubusercontent.com/ML4SCI/ML4SCI_GSoC/main/Messenger/Moon/LPK_Map.csv', header=None)
LPTh = pd.read_csv('https://raw.githubusercontent.com/ML4SCI/ML4SCI_GSoC/main/Messenger/Moon/LPTh_Map.csv', header=None)
LPTi = pd.read_csv('https://raw.githubusercontent.com/ML4SCI/ML4SCI_GSoC/main/Messenger/Moon/LPTi_Map.csv', header=None)
```

```python
# figure of iron map
plt.figure(figsize=(14,7))
sns.heatmap(LPFe, cmap="jet", xticklabels=False, yticklabels=False)
```

```python
# check how they are stored in Dataframe
LPFe.head()
```

```python
# Change the shape from 2-D to 1-D 
```

```python
arr = Albedo.to_numpy()
new_arr = arr.reshape(-1)
```

```python
arr_e = LPFe.to_numpy()
new_e = arr_e.reshape(-1)
```

```python
arr_k = LPK.to_numpy()
new_k = arr_k.reshape(-1)
```

```python
arr_h = LPTh.to_numpy()
new_h = arr_h.reshape(-1)
```

```python
arr_i = LPTi.to_numpy()
new_i = arr_i.reshape(-1)
```

```python
df = pd.DataFrame(data=new_e, columns=['iron'])
```

```python
df['potassium']= new_k
```

```python
df['thorium'] = new_h
df['titanium'] = new_i
df['ALB'] = new_arr
```

```python
df['ALB']
```

```python
# Take the half of map for X_train and y_train data 
df1 = df[:129600]
```

```python
# recontract to one singe DataFrame
df1
```

```python
X = df1[['iron', 'potassium', 'thorium', 'titanium']]
```

```python
y = df1['ALB']
```

```python
# Choosing Linear Regression as Model
from sklearn.linear_model import LinearRegression 
```

```python
lm = LinearRegression()
```

```python
# Train the data
lm.fit(X, y)
```

```python
# Select the other half of the map that our model has not seen
df2 = df[129600:]
df2
```

```python
# Drop column ['ALB'] because this is what we want to Predict
X_tst2 = df2.drop('ALB', axis=1)
```

```python
# 1-D the outputa data shape
pred = lm.predict(X_tst2)
```

```python
# reshape the output of the model to array
pred_2d = np.reshape(pred, (-1,720))
```

```python
# check the format
pred_2d
```

```python
# trasform data from array to pandas dataframe
half = pd.DataFrame(data=pred_2d)
```

```python
# see the result
half
```

```python
# plot the output-predictions
plt.figure(figsize=(12,3))
sns.heatmap(half, cmap="gray", xticklabels=False, yticklabels=False)
```

```python
# see the dufference with the real data
Albedo2 = Albedo[180:] 
```

```python
plt.figure(figsize=(12,3))
sns.heatmap(Albedo2, cmap="gray", xticklabels=False, yticklabels=False)
```

```python
# metrics to calculate how our model perform 
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(Albedo2, half))
print('MSE:', metrics.mean_squared_error(Albedo2, half))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Albedo2, half)))
```

As we see the error in all these metrics is close to 0, 
so that means that out prodicted Model perform Well
